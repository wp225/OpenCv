```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
```

```python
image = cv2.imread('checkerboard.jpeg',0)
plt.imshow(image,c`map='grey')
print(image)
```

```python
plt.imshow(cv2.rectangle(image,[600,600],[800,300]))
```

```python
img_copy = image.copy()
img_copy[0:50,0:20] = 0
plt.imshow(img_copy,cmap='grey')
```

```python
image_cropped = image[200:400,0:250]
plt.imshow(image_cropped,cmap='grey')
```

### Resizing
#### Rescaling

```python
resclaed_image = cv2.resize(image,None,fx=0.5,fy=.5)
plt.imshow(resclaed_image,cmap='grey')
```


#### Resizing to specific

```python
resized_image = cv2.resize(image,dsize=(100,100))
plt.imshow(resized_image,cmap='grey')
```
