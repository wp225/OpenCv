```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image
```

```python
image = cv2.imread('cola.jpeg',1)
plt.imshow(image) #saved as BGR so inverted color 
```

```python
plt.imshow(image[:,:,::-1]) #To show as RGB
```

#### Spliting and Merging 

```python
b,g,r = cv2.split(image)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(141)
plt.imshow(b,cmap='grey')
plt.title('blue')
plt.subplot(142)
plt.imshow(g,cmap='grey')
plt.title('green')
plt.subplot(143)
plt.imshow(r,cmap='grey')
plt.title('red')
plt.subplot(144)
plt.imshow(cv2.merge((r,g,b)))
plt.title('merged')
plt.tight_layout()
plt.show()
```

#### convert between color space

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
plt.figure(figsize=(12,8))
image_hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
for i,img in enumerate(cv2.split(image_hsv)):
    plt.subplot(1,3,i+1)
    plt.imshow(img,cmap='grey')

plt.tight_layout()
plt.show()
```

```python
hsv_image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(hsv_image)
h_new=h-10
hsv_image = cv2.merge((h_new,s,v))
```

```python
plt.imshow(cv2.cvtColor(hsv_image,cv2.COLOR_HSV2RGB))
```

```python
hsv_changed_image = cv2.cvtColor(hsv_image,cv2.COLOR_HSV2RGB)
```

```python
cv2.imwrite('cola_h_changed.jpg',hsv_changed_image)
```

```python
window = cv2.namedWindow('w1')
cv2.imshow(window,hsv_changed_image)
cv2.waitKey(2)
cv2.destroyWindow(window)
```
